# Visualizing Wealth: An Exploration of the World's Billionaires
--------------------------------------------------------------------------------------------------------

<p style="font-size: 17px;">This Python notebook showcases three data visualizations created in Tableau using my "Leading Industries Dataset". The visualizations - "The World's Billionaires", "Industry Trend", and "Billionaires Entering / Exiting" - provide a comprehensive analysis of billionaire wealth and trends across the world.</p>

<p style="font-size: 17px;">"The World's Billionaires" shows the distribution of billionaire wealth across countries and regions, "Industry Trend" analyzes the rise and fall of wealth across different industries, and "Billionaires Entering / Exiting" examines the impact of new entrants and exits on the global market.</p>

<p style="font-size: 17px;">This notebook highlights the power of data visualization tools like Tableau in gaining insights into complex datasets.</p>
